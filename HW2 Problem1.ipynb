```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

The first function we will be looking at is 
z = cos($\sqrt{x^2 + y^2})$

```
# Function 1
def f1(x, y):
  return np.cos(np.sqrt(x**2 + y**2))

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
plt.show()
```

With this contour plot, we can see that there is a local maxima at the point (0, 0)

```
# Point 1
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();

X1 = np.array([0])
Y1 = np.array([0])
plt.plot(X1, Y1, "ro")
print(f1(X1,Y1))
```

The second function we will be looking at is z = x^2 + y^2

```
# Function 2
def f2(x, y):
  return x**2 + y**2

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
plt.show()
```

Using the 3D plot and the contour plot, we can see a global minimum occurs at the point (0, 0)

```
# Point 2
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();

X1 = np.array([0])
Y1 = np.array([0])
plt.plot(X1, Y1, "ro")
print(f1(X1,Y1))
```

The third and final function we will be looking at is z = tan(${x/y}$)

```
# Function 3
def f3(x, y):
  return np.tan(x/y)
  # return 1 / (x**2 + y**2)
  # return np.sin(np.sqrt(x ** 2 + y ** 2))
    # return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
# ax.view_init(80, 35)
plt.show()
```

Using the contour plot, we can see a gradient of 0 along the line (x = 0), so any value of Y can be used here

```
# Point 3
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();

X1 = np.array([0])
Y1 = np.array([3])
plt.plot(X1, Y1, "ro")
print(f3(X1,Y1))
```
