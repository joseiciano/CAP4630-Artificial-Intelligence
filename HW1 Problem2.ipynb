Outputs the product of a list of numpy arrays. If we cannot take the product of two matrices, prints out their dimensions. 


```
import numpy as np
```

Matrix multiplication between two matrices, A and B, is only possible if A is a (M x N) dimensional matrix and B is a (N x K) dimensional matrix, where M, N, and K are any positive number.

Example:  
Let A =  
$[1, 2, 3]$    
$[4, 5, 6]$  
$[7, 8, 9]$

Let B =   
$[4]$  
$[1]$

A*B would not be possible since A is a (3 x 3) matrix and B is a (2 x 1) Matrix. 

Example:  
Let A =  
$[1, 2, 3]$    
$[4, 5, 6]$  
$[7, 8, 9]$

Let B =   
$[1, 2]$    
$[4, 1]$  
$[7, 8]$

A * B is possible since A is a (3 x 3) matrix and B is a (3 x 2) matrix


Also, note that matrix multiplication is not commutative since it relies on matrix dimensions. Thus, A * B is not the same as B * A


```
def multiply_matrices(matrices):
  if matrices is None or len(matrices) == 0:
    try:
      raise Exception("No matrices were passed. Need at least two to multiply")
    except Exception as e:
      print(e)
  elif len(matrices) == 1:
    try:
      raise Exception("Only one matrix was passed. Need at least two to multiply")
    except Exception as e:
      print(e)
  else:
    for i in range(len(matrices)):

      # Get the first matrix, A
      current_matrix = matrices[i]
      
      # Start from 0 since A * B is not the same as B * A, so we could have different things
      for j in range(len(matrices)):
        if i != j:
          try:
            print(f"{current_matrix} * {matrices[j]} = ")
            current_matrix = np.matmul(current_matrix, matrices[j])
            print(current_matrix)
            
          except:
            print(f"Cannot multiply matrices with dimensions {current_matrix.shape} and {matrices[j].shape}")
# multiply_matrices([np.eye(3),np.ones(3)])
# multiply_matrices([]) 
```
